```python
def convert_to_int(x):
    if 'L' in x:
        # Convert "L" (Lakhs) to integer
        return int(float(x.replace('L', '')) * 100000)
    elif 'k' in x:
        # Convert "k" (Thousands) to integer
        return int(float(x.replace('k', '')) * 1000)
    else:
        # If there is no suffix, return as integer
        return int(x)


converted_salaries = [convert_to_int(x) for x in Salaries]
converted_salaries

converted_reviews = [convert_to_int(x) for x in Review]
converted_reviews

converted_jobs = [convert_to_int(x) for x in Job]
converted_jobs

converted_interview = [convert_to_int(x) for x in Interview]
converted_interview

converted_ratings = [(float(y)) for y in rating]
converted_ratings

d = {'name': C_name, 'rating':converted_ratings, 'review':converted_reviews,'salaries':converted_salaries, 'job':converted_jobs, 'interview':converted_interview}

df= pd.DataFrame(d)
df.head(20)
```

```python


def convert_to_int(x):
    if 'L Reviews' in x:
        # Convert "L" (Lakhs) to integer
        return int(float(x.replace('L Reviews', '')) * 100000)
    elif 'L Salaries' in x:
        return int(float(x.replace('L Salaries', '')) * 100000)

    elif 'k Reviews' in x:
        # Convert "k" (Thousands) to integer
        return int(float(x.replace('k Reviews', '')) * 1000)
    elif 'k Salaries' in x:
        # Convert "k" (Thousands) to integer
        return int(float(x.replace('k Salaries', '')) * 1000)
    elif 'Jobs' in x:
        return int(float(x.replace('Jobs', '')))
    elif 'Interviews' in x:
        return int(float(x.replace('Interviews', '')))
    else:
        # If there is no suffix, return as integer
        return int(x)



import numpy as np
from os import read
from typing_extensions import final

final = pd.DataFrame()

for j in range(6,7):
    url = ('https://www.ambitionbox.com/list-of-companies?page={}'.format(j))
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; X64) applewebkit/537.36 (KHTML, like Gecko) chrome/80.0.3987.167 Safari/537.36'}
    webpage = requests.get(url, headers=headers).text

    soup = BeautifulSoup(webpage, 'lxml')

    company = soup.find_all('div', class_ = 'companyCardWrapper')

    C_name =[]
    rating = []
    review = []
    salaries =[]
    job =[]
    interview = []

    for i in company:
        try:
            C_name.append(i.find('h2', class_ = 'companyCardWrapper__companyName').text.strip())
        except:
            C_name.append(np.nan)
        try:
            rating.append(i.find('span', class_ = 'companyCardWrapper__companyRatingValue').text.strip())
        except:
            rating.append(np.nan)
        try:
            review.append(i.find_all('a', class_ = 'companyCardWrapper__ActionWrapper')[0].text.strip())
        except:
            review.append(np.nan)
        try:
            salaries.append(i.find_all('a', class_ = 'companyCardWrapper__ActionWrapper')[1].text.strip())
        except:
            salaries.append(np.nan)
        try:
            interview.append(i.find_all('a', class_ = 'companyCardWrapper__ActionWrapper')[2].text.strip())
        except:
            interview.append(np.nan)
        # try:
        #     job.append(i.find_all('a', class_ = 'companyCardWrapper__ActionWrapper')[3].text.strip())
        # except:
        #     job.append(np.nan)

        # review = [i.replace('Reviews', '') for i in review]
        # salaries = [i.replace('Salaries', '') for i in salaries]
        # job = [i.replace('Jobs', '') for i in job]
        # interview = [i.replace('Interviews', '') for i in interview]


    salaries = [convert_to_int(x) for x in salaries]
    review = [convert_to_int(x) for x in review]
    # job = [convert_to_int(x) for x in job]
    interview = [convert_to_int(x) for x in interview]
    rating = [(float(y)) for y in rating]



    d = {'name': C_name, 'rating':rating, 'review':review,'salaries':salaries, 'interview':interview}

    df= pd.DataFrame(d)

    final= pd.concat([df, final], ignore_index=True)





```
